```python
 #This example uses Python 2.7 and the python-request library.

# from requests import Request, Session
# from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
# import json

# url = 'https://sandbox-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
# parameters = {
#   'start':'1',
#   'limit':'5000',
#   'convert':'USD'
# }
# headers = {
#   'Accepts': 'application/json',
#   'X-CMC_PRO_API_KEY': '247028d7-ad6c-4cf7-823f-cb87a9474127',
# }

# session = Session()
# session.headers.update(headers)

# try:
#   response = session.get(url, params=parameters)
#   data = json.loads(response.text)
#   print(data)
# except (ConnectionError, Timeout, TooManyRedirects) as e:
#   print(e)

from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import pandas as pd

# API URL (Use the PRO API if you have access)
url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'

# API parameters
parameters = {
    'start': '1',
    'limit': '10',  # Fetching only 10 coins for testing; adjust as needed
    'convert': 'USD'
}

# Headers with API key (replace 'YOUR_API_KEY_HERE' with your actual API key)
headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': '247028d7-ad6c-4cf7-823f-cb87a9474127',  
}

# Create a session and update headers
session = Session()
session.headers.update(headers)

try:
    # Make the API request
    response = session.get(url, params=parameters)
    data = response.json()
    
    # Extract coin names and other details if data is present
    if 'data' in data:
        coins = [{'id': coin['id'], 'name': coin['name'], 'symbol': coin['symbol'], 'price': coin['quote']['USD']['price']} 
                 for coin in data['data']]
        df = pd.DataFrame(coins)
        print(df)  # Display the data in tabular format
    else:
        print("Unexpected response format:", data)

except (ConnectionError, Timeout, TooManyRedirects) as e:
    print("Error fetching data:", e)
```
